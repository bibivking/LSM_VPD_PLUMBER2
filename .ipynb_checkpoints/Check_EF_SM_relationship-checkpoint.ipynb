```python
model_name      = 'CABLE'
Rain_input      = pd.read_csv(Qle_file,usecols=['site_name', 'obs_Precip'])
EF_hourly_input = pd.read_csv(Qle_file,usecols=['site_name',  model_name+'_EF'])
SM_1_input      = pd.read_csv(SM_1_file,usecols=['site_name', model_name + '_SMtop1m'])
SM_03_input     = pd.read_csv(SM_03_file,usecols=['site_name', model_name + '_SMtop0.3m'])
EF_1day_input   = pd.read_csv(EF_1_day_file,usecols=['site_name', 'model_'+model_name])
EF_5day_input   = pd.read_csv(EF_5_day_file,usecols=['site_name', 'model_'+model_name])
```

```python
site_name       = 'AU-Dry'#'AU-Tum'
mask = (Rain_input['site_name'] == site_name)
```

```python
EF_hourly_tmp   = EF_hourly_input.loc[mask, model_name+'_EF']
SM_1_tmp        = SM_1_input.loc[mask, model_name + '_SMtop1m']
SM_03_tmp       = SM_03_input.loc[mask,model_name + '_SMtop0.3m']
EF_1day_tmp     = EF_1day_input.loc[mask,  'model_'+model_name]
EF_5day_tmp     = EF_5day_input.loc[mask,  'model_'+model_name]
```

```python
EF_1day_tmp[:] = \
    np.where( (EF_1day_tmp.values > 0) & (EF_1day_tmp.values < 1),
               EF_1day_tmp.values, np.nan)  
EF_5day_tmp[:] = \
    np.where( (EF_5day_tmp.values > 0) & (EF_5day_tmp.values < 1),
               EF_5day_tmp.values, np.nan)  
```

```python
EF_hourly   = EF_hourly_tmp.rolling(window=240, min_periods=1).mean()
SM_1        = SM_1_tmp.rolling(window=240, min_periods=1).mean()
SM_03       = SM_03_tmp.rolling(window=240, min_periods=1).mean()
EF_1day     = EF_1day_tmp.rolling(window=240, min_periods=1).mean()
EF_5day     = EF_5day_tmp.rolling(window=240, min_periods=1).mean()
```

```python
np.any(EF_1day<0)
```

```python
# Calculate the percentile for each element in the time series
SM_per_1  = np.array([percentileofscore(SM_1, SM, kind='rank') for SM in SM_1.values])
SM_per_03 = np.array([percentileofscore(SM_03, SM, kind='rank') for SM in SM_03.values])
```

```python
t_s = 0
t_e = -1 #12000
```

```python
# Plotting the original time series and its percentiles on the same figure with different y-axes
fig, ax1 = plt.subplots()

# Plot the original time series on the first y-axis
ax1.plot(SM_per_1[t_s:t_e], color='black', ls='-', label='SM_per_1')
ax1.plot(SM_per_03[t_s:t_e], color='blue', ls='-', label='SM_per_03')
ax1.set_ylabel('SM_percentiles')
ax1.set_xlabel('Time Series', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()

# Plot the percentiles on the second y-axis
ax2.plot(EF_hourly.values[t_s:t_e], color='red', ls='-', label='EF_hourly')
ax2.plot(EF_1day.values[t_s:t_e], color='orange', ls='-', label='EF_1day')
ax2.plot(EF_5day.values[t_s:t_e], color='pink', ls='-', label='EF_5day')
ax2.set_ylabel('EF', color='r')
ax2.set_ylim([0,1])
ax2.tick_params(axis='y', labelcolor='r')

# Adding a legend
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

# Show the plot
plt.show()
```

### SMtop 0.3m

```python
top_thres=80
bot_thres=20
```

```python
SM_input = pd.read_csv(SM_03_file)
mask     = (SM_input['site_name'] == site_name)
SM_tmp   = SM_input[mask]
SM_all       = {}
SM_per_all   = {}
for model_name in model_names:
    if model_name in ['NASAEnt', 'QUINCY','obs']:
        model_name = 'model_mean'
    SM_model = SM_tmp[model_name + '_SMtop0.3m'].values
    SM_all[model_name]     = SM_model
    SM_per_all[model_name] = np.array([percentileofscore(SM_model, SM, kind='rank') for SM in SM_model])
```

```python
for i, model_name in enumerate(model_names):
    if model_name in ['NASAEnt', 'QUINCY', 'obs']:
        model_name_new = 'model_mean'
    else:
        model_name_new = model_name
    
    # If 'model_mean' is supposed to be a mean of multiple models, 
    # you need to calculate it separately before the loop
    
    if i == 0:
        SM_per_bot = pd.DataFrame(np.where(SM_per_all[model_name_new] < bot_thres, SM_per_all[model_name_new], np.nan), columns=[model_name])
        SM_per_top = pd.DataFrame(np.where(SM_per_all[model_name_new] > top_thres, SM_per_all[model_name_new], np.nan), columns=[model_name])
    else:
        SM_per_bot[model_name] = np.where(SM_per_all[model_name_new] < bot_thres, SM_per_all[model_name_new], np.nan)
        SM_per_top[model_name] = np.where(SM_per_all[model_name_new] > top_thres, SM_per_all[model_name_new], np.nan)
```

```python
SM_per_bot[model_names] = SM_per_bot[model_names].where(~SM_per_bot[model_names].isna().any(axis=1), other=np.nan)
SM_per_top[model_names] = SM_per_top[model_names].where(~SM_per_top[model_names].isna().any(axis=1), other=np.nan)
```

```python
total_num = len(SM_per_bot)
num_bot    = np.sum(~np.isnan(SM_per_bot['obs']))
num_top    = np.sum(~np.isnan(SM_per_top['obs']))

print(total_num, bot_thres, 'th:', num_bot, 'accout for', num_bot/total_num*100, top_thres, 'th:', num_top, 'accout for', num_top/total_num*100)
```

```python
SM_03_file     = "/g/data/w97/mm3972/scripts/PLUMBER2/LSM_VPD_PLUMBER2/txt/process1_output/SMtop0.3m_all_sites.csv"
SM_03_input    = pd.read_csv(SM_03_file)
```

```python
site_names, IGBP_types, clim_types, model_names = load_default_list()
model_names    = model_names['model_select_new']
```

```python
means = {}
stds  = {}
names = None

for i, site_name in enumerate(site_names):
    site_mask = (SM_03_input['site_name'] == site_name)
    SM_03_tmp = SM_03_input[site_mask]

    # Select columns that end with '_0.3m'
    columns_to_calculate = [col for col in SM_03_tmp.columns if col.endswith('_SMtop0.3m')]
    
    # Calculate the mean for these columns
    if i == 0:
        names = SM_03_tmp[columns_to_calculate].mean().index

    means[site_name] = SM_03_tmp[columns_to_calculate].mean().values
    stds[site_name]  = SM_03_tmp[columns_to_calculate].std().values
    
```

```python
mean_all_sites = None
for i, site_name in enumerate(site_names):
    if i == 0:
        mean_all_sites = pd.DataFrame(means[site_name][:], columns=[site_name])
        std_all_sites  = pd.DataFrame(stds[site_name][:], columns=[site_name])
    else:
        mean_all_sites.loc[site_name] = means[site_name][:]
        std_all_sites.loc[site_name]  = stds[site_name][:]
```

```python
mean_all = mean_all_sites.T
mean_all.columns = columns_to_calculate
mean_all.to_csv(f'./txt/process1_output/metric_mean_SMtop0.3m_all_sites.csv', index=True)
```

```python
std_all = std_all_sites.T
std_all.columns = columns_to_calculate
std_all.to_csv(f'./txt/process1_output/metric_std_SMtop0.3m_all_sites.csv', index=True)
```

## Calculate the driest sites and wettest sites

```python
driest_sites  = {}
wettest_sites = {}

for model_name in std_all.columns:

    # Sort the DataFrame by column 'A' in descending order
    sorted_mean = mean_all.sort_values(by=model_name, ascending=False)
    wettest_sites[model_name] = sorted_mean.head(40).index.to_list()
    driest_sites[model_name]  = sorted_mean.tail(40).index.to_list()
    
```

## Check the common driest or wettest sites

```python
# Initialize a dictionary to store common elements for each keyword
common_elements = {}

# Iterate over each keyword and its list of sites
for keyword, site_list in driest_sites.items():
    
    # Convert the list of sites to a set for each keyword
    site_set = set(site_list)
    
    # Initialize the intersection with the first set
    intersection = site_set
    
    # Iterate over the rest of the sets and find the intersection
    for other_keyword, other_site_list in driest_sites.items():
        if other_keyword != keyword:  # Skip the current keyword itself
            other_site_set = set(other_site_list)
            intersection = intersection.intersection(other_site_set)
    
    # Convert the intersection set back to a sorted list
    common_elements[keyword] = sorted(list(intersection))

# Print the common elements for each keyword
for keyword, common_sites in common_elements.items():
    print(f"Common sites for '{keyword}': {common_sites}")
```
